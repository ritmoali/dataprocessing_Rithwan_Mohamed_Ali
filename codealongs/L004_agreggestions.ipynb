### Split Apply Commmand
- Select a feature to use as key
- split the dataset into group for each unique key value
- Aply aggregation to each group
- commbine aggregated data into a new dataset

### Group  by
Use pandas .groupby()method to select a key and split into groups
This creates a new DataFrameGroupby object containing the grouped Dtaframe

```python
makes = autos.groupby("make")
```

## Group by multiple colums

```python
autos.groupby(["body-style", "make"])["price"].agg(["min", "max", "mean"])
```

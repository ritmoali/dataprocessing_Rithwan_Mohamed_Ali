## pandas time series
A series or dataframe using a datetime index instead of a rnage index

```python
import pandas as pd 
import numpy as np
```

### creating a datetimeindex
use date_range() with 3 of the 4 parameters 

```python
datatimeindex= pd.date_range(start='2018-01-01', end='2019-12-31', freq="D")
datatimeindex
```

```python
df =pd.DataFrame(
    dict(
        n= range(len(datatimeindex)),
        rand= np.random.random(len(datatimeindex))
    ), index=datatimeindex
)
df
```

```python
df.loc["2018-03-15"]
df.loc["2018-01-15"]
df.loc["2018-03-15":"2018-03-20"]
df.loc["2018-03"]
df.loc["2019"]
df.loc["2019-03-29":"2019-05"]
```

### Resampling 
Resampling is the practice of creating new sample with a lower or higher frequency than the original data.

### Downsampling 
When the sample frequency is lower than the original we can aggregate values from the original data.

```python
from helpers import hdisplay
hdisplay()
df.resample("2D").agg({"n": "max", "rand": "sum" })
```

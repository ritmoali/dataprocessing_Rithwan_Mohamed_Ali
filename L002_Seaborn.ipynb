# Seaborn
- Seaborn is a pyhthon visualization library based on matplotlib. It provides a high level interface for drawing attractive and informative statistical graphics

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", style="smoker", size="size", data=tips)
```
